```python
# Reading from any sql database with username and password
def read_sql_server ():
    import pandas as pd
    import sqlalchemy
    import getpass

    # Define the connection parameters
    server_name = "stone" # Replace with your SQL Server instance name or IP address
    database_name =str(input('insert your database name'))  # Replace with your database name
    username = str(input('please input your Username') ) # Replace with your username
    password = getpass.getpass(prompt="Enter your password: ")  # Input the password securely
    driver ="ODBC Driver 17 for SQL Server"  # Replace with the appropriate ODBC driver

    # Create a connection string with your format
    connection_string = f"mssql+pyodbc://{username}:{password}@{server_name}/{database_name}?driver={driver}"

    # Create a database engine
    engine = sqlalchemy.create_engine(connection_string, fast_executemany=True)

    # Test the connection
    try:
        connection = engine.connect()
        print("Connected to the database.")
    except Exception as e:
        print("Error:", str(e))
    finally:
        if connection:
            connection.close()
            
    sql=input("your sql statement")
    df=pd.read_sql(sql,engine, chunksize=100)
    df1=pd.DataFrame()
    for a in df:
        df1=df1.append(a)
        return (a)
```
