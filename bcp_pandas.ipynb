```python
%%time
import time
import pandas as pd
from bcpandas import SqlCreds, to_sql
import urllib
from sqlalchemy import create_engine

params_source = urllib.parse.quote_plus("DRIVER={SQL Server Native Client 11.0};SERVER=DESKTOP-TS0EFTA;DATABASE=ade;trusted_connection=yes")
engine_source = create_engine("mssql+pyodbc:///?odbc_connect=%s" % params_source)

creds = SqlCreds.from_engine(engine_source)

params_destination = urllib.parse.quote_plus("DRIVER={SQL Server Native Client 11.0};SERVER=DESKTOP-TS0EFTA;DATABASE=sh;trusted_connection=yes")
engine_destination = create_engine("mssql+pyodbc:///?odbc_connect=%s" % params_destination)

creds_dest = SqlCreds.from_engine(engine_destination)

table_names = ['911']
for table_name in table_names:
    sql = f'SELECT * FROM [dbo].[{table_name}]'
    df = pd.read_sql(sql, engine_source,chunksize=50000)
    for chunk in df:

        to_sql(chunk, table_name, creds_dest, if_exists='replace', index=False)
```
